```python
% matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
wordcounts = {}
for line in open("results/word_counts.txt", "r"):
    w, c = line.split()
    wordcounts[w] = int(c)
```

```python
wordcloud = WordCloud(width=500, height=300, max_words=1000,
                      margin=0, colormap="seismic", background_color="white")
```

```python
wordcloud = wordcloud.generate_from_frequencies(wordcounts)
plt.imshow(wordcloud, interpolation='bilinear')
plt.tight_layout(pad=0)
plt.axis("off")
plt.savefig("results/wordcloud.pdf", dpi=600, format="pdf",facecolor="k", bbox_inches="tight")
```
